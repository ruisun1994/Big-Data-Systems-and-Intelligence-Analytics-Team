```python
from __future__ import print_function
import datetime
import xlsxwriter
import urllib3
import unicodedata
from bs4 import BeautifulSoup, SoupStrainer, Comment
import warnings
```

```python
warnings.filterwarnings('ignore')
description_list = []
url_list = []
status_list = []
validTime_list = []
```

```python
url='https://www.w3schools.com/'
```

Send a GET request to the url and wait for a response status. If the status code equals 200, it means the link is valid. 

```python
http = urllib3.PoolManager()
response = http.request('GET',url)
response.status
```

```python
soup=BeautifulSoup(response.data, "lxml")
```

```python
# Crawl all the external links
externals_links = soup.find_all('a')
```

```python
externals_links
```

```python
# for a in soup.find_all('a', href=True):print ("url:",a['href'])
#PartB2 associate the link with a textual description of it from the website
for link in soup.find_all('a'):
    href = link.get('href')
    if "http" in href:
        href = href
    else:
        href = "https://www.w3schools.com/" + link.get('href')
    description = link.string
    description_list.append(description)
    url_list.append(href)
    print(href)
    print(description)
```

```python
# Check whether each link is valid
def check_url(link):
    now = str(datetime.datetime.now())[:19]
    url = "https://www.w3schools.com/" + link.get('href')
    response = http.request('GET',url)
    if response.status != 200:
        status_list.append(0)
        validTime_list.append(now)
    else:
        status_list.append(1)
        validTime_list.append(now)
```

```python
for link in soup.find_all('a'):
        check_url(link)
```

```python
workbook = xlsxwriter.Workbook(u'w3school.xlsx')
worksheet = workbook.add_worksheet(u'space')
title = [u'description', u'url', u'status', u'datetime check']
format_title = workbook.add_format()
format_title.set_border(1)
format_title.set_bg_color('#cccccc')
format_title.set_align('center')
format_title.set_bold()
format_title.set_size(15)
format_body = workbook.add_format()
format_body.set_border()
format_body.set_align('left')
worksheet.set_column('A:B',30)
worksheet.write_row('A1',title,format_title)
worksheet.write_column('A2',description_list,format_body)
worksheet.write_column('B2',url_list,format_body)
worksheet.write_column('C2',status_list,format_body)
worksheet.write_column('D2',validTime_list,format_body)
workbook.close()
```
